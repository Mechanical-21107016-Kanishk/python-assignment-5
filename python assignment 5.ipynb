```python
#3
import math

a = int(input("Enter the length of first side of the triangle: "))
b = int(input("Enter the length of second side of the triangle: "))
c = int(input("Enter the length of third side of the triangle: "))

if a + b > c and b + c > a and c + a > b:
  s = (a + b + c) / 2
  area = math.sqrt(s * (s - a) * (s - b) * (s - c))
  print("Area of the triangle is", area)
else:
  print("The given sides don't form a triangle.")
```

```python
#5
rowNumber = int(input("Enter the number of rows: "))  # user input

i = 0  # variable to repeat the alphabet printing process  = nth line
a = 0    # used to increment argument of chr function to generate next alphabet
line = 1  # variable denoting nth line

for x in range(rowNumber):   # repeating the loop *rowNumber* of times
  i = 0  
  while i < line:
    print(chr(65 + a), end = "")  # printing the alphabet
    i = i + 1  # for printing new alphabet
    a = a + 1  # for new alphabet
  print("")  # for new line
  line = line + 1  # line number incremented
```

```python
#6
upperLimit = int(input("Enter the number till which you want the prime numbers to be printed: "))

for num in range(upperLimit + 1):
   if num > 1:
       for i in range(2, num):
           if (num % i) == 0:
               break
       else:
           print(num)
```

```python
#8 (a)

positiveN = []

for x in range(1, 11):
  userInput = int(input("Enter the input {}: ".format(x)))
  if userInput != 0 and userInput == abs(userInput):
    positiveN.append(userInput)

print("Following are the positive numbers entered", positiveN)
```

```python
#8 (b)

negativeN = []

for x in range(1, 11):
  userInput = int(input("Enter the input {}: ".format(x)))
  if userInput != 0 and userInput == -abs(userInput):
    negativeN.append(userInput)

print("Following are the positive numbers entered", negativeN)
```

```python
#8 (c)

oddN = []

for x in range(1, 11):
  userInput = int(input("Enter the input {}: ".format(x)))
  if userInput != 0 and userInput % 2 == 1:
    oddN.append(userInput)

print("Following are the positive numbers entered", oddN)
```

```python
#8 (e)

N = []

for x in range(1, 11):
  userInput = int(input("Enter the input {}: ".format(x)))
  N.append(userInput)

dictN = {num:N.count(num) for num in N}

print(dictN)
```
